# Swigg Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
s_data = pd.read_csv('Swiggy_Data.csv')
```

```python
s_data
```

**Cleaning the data**

```python
# Dropping NaN values from dataframe

#getting all the NaN rows
nan_df = s_data[s_data.isna().any(axis=1)]
nan_df
```

```python
# dropping NaN values from dataframe
s_data = s_data.dropna(how = 'all')     # 'All' drops only if all values are null
s_data
```

```python
s_data.info()
```

```python
# rename Restaurant Name to Restaurant
s_data.rename(columns = {'Restaurant Name' : 'Restaurant'}, inplace = True)
```

```python
s_data['Rating'].value_counts()
```

```python
def fun(column):
    rating_int = []
    for i in column:
        if i == 'NEW':
            rating_int.append(0)
        elif i == '--':
            rating_int.append(0)
        else:
            rating_int.append(float(i))
    return rating_int
```

```python
s_data['Rating'] = fun(s_data['Rating'])
```

```python
s_data['Rating']
```

```python
s_data['Rating'].replace(0.0,s_data['Rating'].mean(),inplace=True)
```

```python
s_data['Rating'] = round(s_data['Rating'],1)
```

```python
s_data['Rating']
```

```python
#Restaurant with maximum rating at paticular place

s_data = s_data[s_data['Location']=='Bangalore']
s_data.groupby('Restaurant')['Rating'].max().sort_values(ascending=False).head(1).reset_index()
```

```python
s_data
```

```python
# rename Average Price to  Average Price(cost for 2)
s_data.rename(columns = {'Average Price' : 'Average Price(cost for 2)'},inplace=True)
```

```python
# Replacing average price by just integer values
s_data['Average Price(cost for 2)'] = s_data['Average Price(cost for 2)'].str.extract('(\d+)').fillna(0).astype(int)
```

```python
s_data
```

```python
s_data.info()
```

```python
#cleaning number of rating values

s_data['Number of Ratings'] = s_data['Number of Ratings'].replace('Too Few Ratings',0)
```

```python
s_data['Number of Ratings'] = s_data['Number of Ratings'].str.replace(r'\+.*','',regex= True)
```

```python
s_data['Number of Ratings']
```

```python
#cleaning Null values

s_data['Number of Ratings'] = s_data['Number of Ratings'].str.replace(r'\D','',regex= True)
```

```python
s_data['Number of Ratings'] = s_data['Number of Ratings'].replace(np.nan,0).astype(int)
```

```python
s_data.info()
```

```python
s_data['Offer Name'] = s_data['Offer Name'].str.replace('\n',',',regex = True)
```

```python
s_data['Offer Name'].fillna('No Info',inplace = True)
```

```python
s_data.info()
```

```python
# rename Pure Veg to  Veg Or Non-veg
s_data.rename(columns = {'Pure Veg' : 'Veg or Non-veg'}, inplace = True)
```

```python
s_data['Veg or Non-veg'] = s_data['Veg or Non-veg'].replace('Yes', 'Veg')
s_data['Veg or Non-veg'] = s_data['Veg or Non-veg'].replace('No', 'Non-veg')
```

```python
s_data_reset = s_data.reset_index(drop=True)
```

```python
s_data_reset
```

```python
s_data_reset.to_csv('Swiggy-Data.csv')
```

```python
# 1. Restaurant with Maximum Rating in Bangalore
max_rating_in_Bangalore = s_data_reset.groupby('Restaurant')['Rating'].max().sort_values(ascending=False).head(5) 
max_rating_in_Bangalore
```

```python
plt.figure(figsize=(5, 4))
max_rating_in_Bangalore.plot.bar(color='red')
plt.ylabel("Rating in number")
plt.xlabel("Restaurent Name")
plt.title("Restaurant with Maximum Rating in Bangalore") 
plt.show()
```

```python
# Price for two members in different restaurant
plt.figure(figsize = (10,6))
s_data_reset.groupby(['Restaurant'])['Average Price(cost for 2)'].mean().sort_values(ascending = False)[:10].plot(kind='pie',autopct = '%.2f%%')
plt.xlabel('Restaurant')
plt.ylabel('Price for 2')
plt.show()
```

```python
#Mostly availabel Cuisine
plt.figure(figsize=(8,6))
sns.set_style('dark')
pd.Series(s_data_reset['Cuisine'].value_counts(ascending = False)[:20].plot(kind='barh',color = 'darkblue'))
plt.title('Most available Cuisine')
plt.xlabel('Restaurant')
plt.ylabel('Cuisine')
plt.show()
```

```python
#Highest Rating
plt.figure(figsize=(8,6))
pd.Series(s_data_reset['Rating'].value_counts(ascending = False).head(10).plot(kind='pie',autopct= '%.2f%%'))
plt.show()
```

```python
#Outlets of restaurant
plt.figure(figsize = (8,9))
sns.color_palette('gray')
sns.set_style('white')
pd.Series(s_data_reset['Restaurant'].value_counts(ascending=False)[:10].plot(kind='barh',color='grey'))
plt.title('Most Apperance Restaurant')
plt.xlabel('Apperance')
plt.ylabel('Restaurant')
plt.show()
```
